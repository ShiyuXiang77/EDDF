```python
#取该文件夹下，取Cipher_renellm_mix_prompt_data_our_work.json和flip_renellm_adv100-200_our_work.json的前200个元素，unsafe_Template_rule_prompt(1).json的后200个元素，重新生成三个json文件
import os
import json

def process_json_files(folder_path):
    # 定义三个 JSON 文件的路径
    file1 = os.path.join(folder_path, "Cipher_renellm_mix_prompt_data_our_work.json")
    file2 = os.path.join(folder_path, "flip_renellm_adv100-200_our_work.json")
    file3 = os.path.join(folder_path, "unsafe_Template_rule_prompt(1).json")
    
    # 处理第一个文件：提取前200个元素
    with open(file1, 'r', encoding='utf-8') as f1:
        data1 = json.load(f1)
    first_200_file1 = data1[:200]
    
    # 处理第二个文件：提取前200个元素
    with open(file2, 'r', encoding='utf-8') as f2:
        data2 = json.load(f2)
    first_200_file2 = data2[:200]
    
    # 处理第三个文件：提取后200个元素
    with open(file3, 'r', encoding='utf-8') as f3:
        data3 = json.load(f3)
    last_200_file3 = data3[-200:]
    
    # 定义新的输出文件路径
    output1 = os.path.join(folder_path, "new_Cipher_renellm_mix_prompt_data_our_work.json")
    output2 = os.path.join(folder_path, "new_flip_renellm_adv100-200_our_work.json")
    output3 = os.path.join(folder_path, "new_unsafe_Template_rule_prompt(1).json")
    
    # 将提取的数据分别保存到新的 JSON 文件中
    with open(output1, 'w', encoding='utf-8') as out1:
        json.dump(first_200_file1, out1, ensure_ascii=False, indent=4)
    with open(output2, 'w', encoding='utf-8') as out2:
        json.dump(first_200_file2, out2, ensure_ascii=False, indent=4)
    with open(output3, 'w', encoding='utf-8') as out3:
        json.dump(last_200_file3, out3, ensure_ascii=False, indent=4)
    
    print("新生成的 JSON 文件已保存。")

# 使用示例
folder_path = "../mix"  # 请替换为你的文件夹路径
process_json_files(folder_path)
```

```python
import json
import random

def shuffle_json_file(input_file, output_file):
    # 读取 JSON 文件
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # 确保数据为列表，否则无法打乱顺序
    if isinstance(data, list):
        random.shuffle(data)
    else:
        print("JSON 文件内容不是列表，无法打乱顺序。")
        return
    
    # 将打乱后的数据写入新的 JSON 文件
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    print(f"已将打乱后的内容保存到 {output_file}")

# 示例调用，请替换实际文件路径
input_file = "unsafe_Cipher_prompt_data.json"
output_file = "unsafe_Cipher_prompt_shuffled.json"
shuffle_json_file(input_file, output_file)
```

```python
import json
import random

def filter_and_select(data, num_samples=36):
    """筛选出 'is harmful' 为 true 的元素，并从 'is harmful' 为 false 的元素中随机选取指定数量的元素"""
    harmful_items = [item for item in data if item.get("is harmful") == True]
    non_harmful_items = [item for item in data if item.get("is harmful") == False]

    if len(non_harmful_items) < num_samples:
        raise ValueError("数据中 'is harmful' 为 false 的元素不足以选取指定数量")
    
    # 随机选取指定数量的非有害元素
    selected_non_harmful_items = random.sample(non_harmful_items, num_samples)
    
    # 将有害元素和选中的非有害元素拼接在一起
    selected_data = harmful_items + selected_non_harmful_items
    return selected_data

def main():
    # 读取文件
    with open('cipher.json', 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # 筛选并选取 36 个非有害元素
    selected_data = filter_and_select(data, 36)
    
    # 保存选中的元素到新的文件
    with open('cipher_final_data.json', 'w', encoding='utf-8') as f:
        json.dump(selected_data, f, ensure_ascii=False, indent=4)
    
    print(f"成功选取并保存了 {len(selected_data)} 个元素到 'selected_combined_data.json'")

if __name__ == "__main__":
    main()
```

```python
#将文件夹下json文件中“is harmful”为true的筛选出来组成一个新json文件，原文件中该元素删除

import os
import json

def filter_harmful_json_files(input_folder, output_folder):
    # 遍历文件夹中的所有 JSON 文件
    for filename in os.listdir(input_folder):
        if filename.endswith(".json"):
            file_path = os.path.join(input_folder, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
                
            # 筛选出 'is_harmful' 为 True 的元素
            harmful_data = [item for item in data if item.get("is_harmful") == True]
            
            # 更新原文件内容，删除 'is_harmful' 为 True 的元素
            filtered_data = [item for item in data if item.get("is_harmful") != True]

            # 将修改后的数据写回原文件
            with open(file_path, 'w', encoding='utf-8') as file:
                json.dump(filtered_data, file, ensure_ascii=False, indent=4)
            
            # 如果有 harmful 数据，保存为新的文件
            if harmful_data:
                output_file_path = os.path.join(output_folder, f"harmful_{filename}")
                with open(output_file_path, 'w', encoding='utf-8') as output_file:
                    json.dump(harmful_data, output_file, ensure_ascii=False, indent=4)
                
                print(f"Processed {filename}, harmful data saved to {output_file_path}")

# 使用示例
input_folder = '../mix'  # 输入文件夹路径
output_folder = './chpher.json'  # 输出文件夹路径

filter_harmful_json_files(input_folder, output_folder)
```
